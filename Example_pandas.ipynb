Pandas - обязательная часть жизненного цикла Data science. Это самая популярная и широко используемая библиотека Python для науки о данных, наряду с NumPy, matplotlib. Pandas предоставляет быстрые и гибкие структуры данных, такие как dataframe и series, которые предназначены для очень простой и интуитивно понятной работы со структурированными данными.

Содержит:


*   Богатые функциональные возможности
*   Высокоуровневые структуры данных и инструменты манипулирования
*   ETL (извлечение, преобразование, загрузка) для преобразования и хранения данных. Имеет поддержку загрузки CSV-файлов, TXT-файлов, XLSX-файлов

```python
# еще один способ фильтрации -- это метод query()
data.query('A > 5 and B < 48') #команда для фильтрации пишется в ковычках в виде строки, в pandas можно писать sql запросы
```

```python
S = pd.Series(['Мышка','Нарушка'])
print(S)
```

```python
print(S.index) #индексы
print(list(S.index)) #индексы
print(S.values) #список значений
```

```python
#индексы можно задать ручками
new_s = pd.Series(S.values, index=['1','2'])
new_s['3']='Квакушка' #добавление нового значения
new_s['4']='Квакушка'
print(new_s)
```
